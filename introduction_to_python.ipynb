Operator in python
1. Arith matic : + - * / %
2. Comparison : > >= < <= == !=
3. Logical operator:  and or

```python
x = 10; y = 2 #if multiple input in a single line that time we should use semicolon

sum = x + y #sum
sub = x + y #sub
mul = x * y #mul
div = x / y #div
rem = x % y #mod

print('sum = ',sum, ' sub = ',sub, " mul = ",mul," div = ",div," rem = ",rem)
```

```python
x = 10; y = 3

greater_than = x >3

greater_than_equal = x>=3
print(greater_than, greater_than_equal)

less_than = x <3

less_than_equal = x<=3
print(less_than, less_than_equal)

equal =  x==y
not_equal = x!=y
print(equal,not_equal)
```

```python
#logical operator
x = 5; y = 3 ; z = 2
result = (x<y) and (x<z) #if x is smaller than y and also x is less than z than the result will be true
print(result)

result2 = (x>y) or (x>z) #if x is greater than y or x is greater than z than the result will be True
print(result2)
```

```python
#while loop
count = 1
while count <10:
    print(count,'Hello')
    count = count+1
print(count)
```

```python
sum = 0

for i in range(1,11):
    sum+=i
    print(i,'is added to the sum')
print(sum)
```

```python
#sum of 1 to n except even numbers
sum = 0

for i in range(1,11):
    print(i,'is processing')
    if i%2 ==0:#even
        continue
    sum+=i
    print(i,'is added to the sum')
print(sum)
```
